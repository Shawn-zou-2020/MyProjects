```python
# Load packages
import pandas as pd
import numpy as np
from datetime import datetime as dt
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
# Load the data
etfs_all = pd.read_feather('etfdata_new.feather')
```

# Data Manipulation

```python
# Only need the data of ETFs, SHRCD=73
etfs = etfs_all.loc[etfs_all['SHRCD']==73]

# Add annual risk-free ret
irx = pd.read_csv('^IRX_.csv',parse_dates=[0], index_col=[0])[['Close']]

# Join
etfs = etfs.join(irx, how = 'left', on = 'DATE')
etfs = etfs.rename(columns={'Close':'ARRET'})
etfs.head()

# Fillna
etfs['ARRET'] = etfs['ARRET'].fillna(method='ffill')

# N_t = (date-date.shift()).dt.days
etfs = etfs.set_index(['PERMNO'])
etfs['LAGD'] = etfs[['DATE']].groupby('PERMNO').shift()
etfs['N_t'] = (etfs['DATE'] - etfs['LAGD']).dt.days

# Risk return 
etfs['RRET'] = etfs['ARRET'].shift() * etfs['N_t']/36500

# Excess return 
etfs['EXRET'] = etfs['RET'] - etfs['RRET'] 
```

**Time period**
- We choose the 5-year data from 2014 to 2018 to backtest our investment strategy because we want to avoid the impact of Covid 19 period and create a strategy based on the period before covid hit the economy. The data from 2014-2018 reflects fewer extreme strikes in volatility and can better validate our strategy.
- We adopt moving average metrics in our 5-year analysis.

```python
# Time Periods data slicing
etfs_a = etfs.loc[(etfs['SHRCD']==73)&(etfs['DATE']<'2019-01-01')&(etfs['DATE']>='2013-01-01')].copy()
```

# Choose ETFs

**ETFs Choosing strategy**
- We choose all the ETFs from the dataset and adjust our portfolio every day based on historical performance.
- We use 66 days (3-month trading days) of data to define moving average/moving standard deviation metrics.
- We use lagged terms of all the metrics in case of data leakage.
- Criteria: ETFs with good return and high volatility; using high return to control ETFs' quality and using high volatility to make reversal strategy work better.


```python
# MKTP: Market capitalization = price * shares outstanding
# MAMKTP: Moving average of market capitalization of 66 trading days
etfs_a['MKTP'] = etfs_a['PRC'] * etfs_a['SHROUT']
etfs_a['MAMKTP'] = etfs_a.groupby('PERMNO')['MKTP'].apply(lambda x: x.rolling(66).mean())

# MSTD: Standard deviation of excess return of 66 trading days
# MARET: Moving average of excess return of 66 trading days
# MARS: Moving average of MARET * MSTD of 66 trading days
etfs_a['MSTD'] = etfs_a[['EXRET']].groupby('PERMNO').apply(lambda x: x.rolling(66).std())
etfs_a['MARET'] = etfs_a[['EXRET']].groupby('PERMNO').apply(lambda x: x.rolling(66).mean())
etfs_a['MARS'] = etfs_a['MARET']*etfs_a['MSTD']

# LAGMKTP: Lag 1 day of MKTP variable
# LAGMARS: Lag 1 day of MARS variable
etfs_a['LAGMKTP'] = etfs_a['MAMKTP'].groupby('PERMNO').shift()
etfs_a['LAGMARS'] = etfs_a['MARS'].groupby('PERMNO').shift()
```

**Data Cleaning**
- We only keep the data after 2014.
- We only keep small-cap ETFs with the market capital less than 2 billions because reversal strategy usually works better in small-cap ETFs. we set 2,000,000 as market cap limit here because the unit is in thousands in this dataset. 
- We drop Nan generated by lagging and rolling;

```python
etfs_a = etfs_a.reset_index()
etfs_a = etfs_a.set_index('DATE')
etfs1 = etfs_a.loc['2014-01-01':].copy()
etfs1 = etfs1.loc[etfs1['LAGMKTP']<=2000000].copy()
etfs1 = etfs1.dropna(subset=['LAGMKTP','LAGMARS'])
```

**Filter the ETFs**
- We choose Top 50 RS(return * volatility) ETFs because we want to choose ETF with high volatility and high return.

```python
etfs_b = etfs1.reset_index()
etfs_b = etfs_b.set_index(['DATE','PERMNO'])[['LAGMKTP','LAGMARS']]

# Choose Top 50 RS(return * volatility) ETFs
etfs_b['rank_mars']=etfs_b['LAGMARS'].groupby('DATE').rank(ascending=False)
etfs_b=etfs_b.sort_values(by=['DATE','rank_mars'], ascending=True).groupby('DATE').head(50)

etfs_b
```

**Data manipulation**
- Only keep ETFs satisfying the conditions above.

```python
etf_list = etfs_b.index.to_list()

etfs_c = etfs1.reset_index()
etfs_c = etfs_c.set_index(['DATE','PERMNO'])
# Only get the data of the chosen etfs
etf2 = etfs_c.loc[etfs_c.index.isin(etf_list)].copy()
```

```python
# ret the index for further convenience
etf2 = etf2.reset_index()
etf2 = etf2.set_index(['DATE','PERMNO'])
```

# Reversal Strategy

**Reversal Strategy**
- We create a DIFF variable as the ranking signal. The DIFF variable is the difference between lagged 5-day moving average return and lag 1-day return.
- DIFF variable is the signal for a sudden drop or a sudden rise;
- Rank the DIFF variable into 20 groups, short rank20 by 100% and long rank1 by 100%.

```python
# Some abandoned ideas...
# etf2['MA2'] = etf2.groupby('PERMNO')[['RET']].apply(lambda x: x.rolling(2).mean())
# etf2['M5STD'] = etf2.groupby('PERMNO')[['EXRET']].apply(lambda x: x.rolling(5).std())
# etf2['M5SR'] = etf2['MA5']/etf2['M5STD']
# etf2['LAGM5SR'] = etf2.groupby('PERMNO')[['M5SR']].shift()


# MA5: 5-day moving average return
# LAGMA5: lag 1 day of MA5
# LAGRET: lag 1 day of RET
# DIFF: LAGRET - MA5
etf2['MA5'] = etf2.groupby('PERMNO')[['RET']].apply(lambda x: x.rolling(5).mean())
etf2['LAGMA5'] = etf2.groupby('PERMNO')[['MA5']].shift()
etf2['LAGRET'] = etf2.groupby('PERMNO')[['RET']].shift()
etf2['DIFF'] = etf2['LAGRET'] - etf2['LAGMA5']

etf2 = etf2.dropna(subset=['LAGMA5','MA5','LAGRET'])
etf2.head()
```

- **Rank all etfs based on LAGMA5 into 20 groups**

```python
etf2 = etf2.reset_index()
etf2 = etf2.set_index(['DATE','PERMNO'])
etf2['RANK'] = etf2['DIFF'].groupby(['DATE']).apply(lambda x: pd.qcut(x, 20, labels=range(1,21)))
```

- **We observe return of current period to evaluate the strategy's effectiveness**

```python
etfports = etf2.groupby(['RANK','DATE'])[['RET']].mean()
etfports.groupby(['RANK']).describe()
```

**Operation**
- Long the rank1 and short the rank20

```python
lmh = etfports.loc[1] - etfports.loc[20]
stats = lmh.describe()
stats
```

```python
m = stats.loc['mean']
sd = stats.loc['std']
n = stats.loc['count']
```

```python
# Average daily return
m
```

```python
# T-test
se = sd/np.sqrt(n)
t = m/se
t
```

```python
# Sharp ratio
sr = m/sd*np.sqrt(264)
sr
```

- **Plot the return patterns of reversal stragety and B&H startegy on chosen ETFs**

```python
etf3 = etf2.copy()
etf3 = etf3.reset_index()
etf3 = etf3.set_index('DATE')
etf3['CC'] = [1 if i==1 else -1 if i==20 else 1 for i in etf3['RANK']]
etf3['PORT'] = etf3['CC']*etf3['RET']
etf3_plot = etf3[['RET','PORT']].groupby('DATE').mean()
etf3_plot['CUMRET'] = (1+etf3_plot['RET']).cumprod()
etf3_plot['CUMRETP'] = (1+etf3_plot['PORT']).cumprod()

from plotnine import *
etf3_plot['DATE'] = etf3_plot.index
(ggplot(etf3_plot,aes('DATE','CUMRETP'))
+ geom_line(color='Red')
+ geom_point(color='Red')
+ geom_line(aes('DATE','CUMRET'))
+ geom_point(aes('DATE','CUMRET'))
+ theme_classic()
+ theme(figure_size=(15,5)))
```

**Is the reversal relationship strong between excess return and DIFF variable?**
- We run the for loop regressions to observe the coefficient of DIFF.

```python
allparams = etf2.groupby(['DATE']).apply(lambda x: smf.ols('EXRET ~ DIFF', data=x).fit().params)
```
