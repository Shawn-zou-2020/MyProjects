```python
dfc['date'] = [i[0] for i in dfc['Scrape File'].str.findall("\d+_\d+")]
```

```python
dfc['date'] = pd.to_datetime(dfc['date'])
```

```python
dfc = dfc.set_index(['id','date'])
dfc['lm_review'] =  dfc['number_of_reviews'].groupby(['id']).shift()
# dfc['month_review'] = dfc['number_of_reviews'] - dfc['lm_review']
dfc = dfc.sort_index()
dfc.head(20)
```
